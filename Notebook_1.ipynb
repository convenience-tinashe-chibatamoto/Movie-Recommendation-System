<h3>
Calculate a weighted rating
</h3>

WR = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C
<br><br>
**v - number of votes for a movie<br>
m - minimum number of votes required<br>
R - average rating of the movie<br>
C - average rating across all movies<br><br>**

*This is the weighted rating formula used by IMDb for their movie rating system.*

```python
m = movies["vote_count"].quantile(0.9)
m
```

```python
C = movies["vote_average"].mean()
C
```

WR = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C

```python
movies_filtered = movies.copy().loc[movies["vote_count"] >= m]
movies_filtered
```

```python
def weighted_rating(df, m=m, C=C):
    R = df["vote_average"]
    v = df["vote_count"]
    wr = ((v / v+m) * R) + (m / (v+m) * C)
    return wr
```

```python
movies_filtered["weighted_rating"] = movies_filtered.apply(weighted_rating, axis=1)
movies_filtered.tail()
```

```python
movies_filtered.sort_values("weighted_rating", ascending=False)[["title", "weighted_rating"]].head(10)
```
